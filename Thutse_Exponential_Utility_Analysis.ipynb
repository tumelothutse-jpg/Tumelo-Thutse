```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters from Section 4.1
a1 = 0.6
sigma1, sigma2 = 0.4, 0.3
lambda1 = 1.4
lambda2 = 1.2
gamma = 0.5

# Rho grid
N = 200
rho_seq = np.linspace(-1, 1, N)

# Pi_1 as a function of rho
Pi_1 = np.where((np.isclose(rho_seq, 1)) | (np.isclose(rho_seq, -1)),
               np.inf, lambda1 * sigma2**2 - lambda2 * rho_seq * sigma1 * sigma2) / (
    gamma * sigma1**2 * sigma2**2 * (1 - rho_seq**2)
)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(rho_seq, Pi_1, label=r"$\pi^{*}_1(\rho)\quad \rho\in(-1,1)$")
plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel(r"$\rho$")
plt.ylabel(r"$\pi^{*}_1$")
plt.title(r"Dependence of $\pi^{*}_1$ on $\rho$")
plt.legend()
#plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sigma2 = 0.3
lambda1 = 1.4
lambda2 = 1.2
gamma = 0.5

# Sigma1 grid
N = 200
sigma1 = np.linspace(0.01, 0.9, N)

# Rho values to compare
#rho_values = np.linspace(-1, 1, N)
rho_values = [0.5, 0.6, 0.7, -0.5, -0.6, -0.7]

# Plot
plt.figure(figsize=(8, 5))
for rho in rho_values:
    numer = -lambda2 * sigma1**2 + lambda1 * rho * sigma1 * sigma2
    denom = gamma * sigma1**2 * sigma2**2 * (1 - rho**2)
    Pi_2 = numer / denom

    plt.plot(sigma1, Pi_2, label=fr"$\pi^*_2(\sigma_1), \rho={rho}$")

plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel(r"$\sigma_1$")
plt.ylabel(r"$\pi^{*}_2$")
plt.title(r"Dependence of $\pi^{*}_2$ on $\sigma_1$ for different $\rho$")
plt.legend()
#plt.grid(True)
plt.show()
```
