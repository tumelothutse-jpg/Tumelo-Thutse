```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 0.6
sigma1, sigma2 = 0.4, 0.3
lambda1 = 1.4
lambda2 = 1.2
gamma = 0.5

# Grids
N = 200
rho_seq = [0.3, -0.3, 0.5, -0.5, 0.7, -0.7]  # fixed correlation values
x_values = np.linspace(0.01, 10000, N)  # reserve levels

plt.figure(figsize=(8, 5))

for rho in rho_seq:
    # slope for given rho
    numer = lambda1 * sigma2**2 - lambda2 * rho * sigma1 * sigma2
    denom = alpha * sigma1**2 * sigma2**2 * (1 - rho**2)
    slope = np.inf if np.isclose(denom, 0) else numer / denom

    # full curve in x
    Pi_1 = slope * x_values
    plt.plot(x_values, Pi_1, label=fr"$\rho={rho}$")

plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel(r"Reserve $x$")
plt.ylabel(r"$\pi^{*}_1(x)$")
plt.title(r"Dependence of $\pi^{*}_1$ on $x$ for different $\rho$ (with $\alpha = 0.6$)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sigma2 = 0.3
lambda1 = 1.4
lambda2 = 1.2
gamma = 0.5

# Sigma1 grid
N = 200
sigma1 = np.linspace(0.01, 0.9, N)

# Values to compare
rho_values = [0.5, 0.6, 0.7, -0.5, -0.6, -0.7]
x_values = [1, 2.5, 5.0]

# Plot
plt.figure(figsize=(10, 6))
for x in x_values:
    for rho in rho_values:
        numer = -lambda2 * sigma1**2 + lambda1 * rho * sigma1 * sigma2
        denom = gamma * sigma1**2 * sigma2**2 * (1 - rho**2)
        Pi_2 = (numer / denom) * x
        plt.plot(sigma1, Pi_2, label=fr"$x={x}, \rho={rho}$")

plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel(r"$\sigma_1$")
plt.ylabel(r"$\pi^{*}_2(\sigma_1)$")
plt.title(r"Dependence of $\pi^{*}_2$ on $\sigma_1$ for different $\rho$ and $x$")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
#plt.grid(True)
plt.tight_layout()
plt.show()
```
