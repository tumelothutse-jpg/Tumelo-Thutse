```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a1 = 0.6
sigma1, sigma2 = 0.4, 0.3
lambda1 = 1.4
lambda2 = 1.2
gamma = 0.5

# Grids
N = 200
rho_seq = np.linspace(-0.99, 0.99, N)  # avoid division by zero at ±1
x_values = [1, 2.5, 5.0]

# Base slope (independent of x)
numer = lambda1 * sigma2**2 - lambda2 * rho_seq * sigma1 * sigma2
denom = sigma1**2 * sigma2**2 * (1 - rho_seq**2)
slope = np.where(np.isclose(denom, 0), np.inf, numer / denom)

# Plot for each x
plt.figure(figsize=(8, 5))
for x in x_values:
    Pi_1 = slope * x
    plt.plot(rho_seq, Pi_1, label=fr"$x={x}$")

plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel(r"Correlation $\rho$")
plt.ylabel(r"$\pi^{*}_1(\rho,x)$")
plt.title(r"Dependence of $\pi^{*}_1$ on $\rho$ for different reserves $x$")
plt.legend()
#plt.grid(True)
plt.show()



```
